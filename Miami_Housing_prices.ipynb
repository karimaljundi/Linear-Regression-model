```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
house_data = pd.read_csv('miami-housing.csv')
```

### Data Prep

```python
y = house_data['SALE_PRC']
```

```python
X = house_data.drop('SALE_PRC', axis=1)
X
```

### Data Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

model  = LinearRegression()

model.fit(X_train, y_train)
```

### Predict values from Linear Regression model

```python
y_model_train_pred = model.predict(X_train)
y_model_test_pred = model.predict(X_test)
```

```python
print(y_model_train_pred)
```

```python
print(y_model_test_pred)
```

### Evaluate models to verify predictions

```python
y_train
```

```python
y_model_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

model_train_mse = mean_squared_error(y_train, y_model_train_pred)
model_train_r2 = r2_score(y_train, y_model_train_pred)
model_test_mse = mean_squared_error(y_test, y_model_test_pred)
model_test_r2 = r2_score(y_test, y_model_test_pred)
```

```python
model_results = pd.DataFrame(['Linear Regression', model_test_mse, model_train_r2, model_test_mse, model_test_r2]).transpose()
model_results.columns = ['Method', 'Training MSE', 'Training R2', 'Testing MSE', 'Testing R2']
```

```python
model_results
```

## **Random Forest**

## **Train the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

## **Evaluate the model**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_test_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Testing MSE', 'Testing R2']
rf_results
```

# **Model Compare**

```python
df_model = pd.concat([model_results, rf_results], axis=0)
df_model
```
